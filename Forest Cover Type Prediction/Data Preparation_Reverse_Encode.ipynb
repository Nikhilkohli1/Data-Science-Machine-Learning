```python
cont_data=forest.loc[:,'Elevation':'Horizontal_Distance_To_Fire_Points']

binary_data=forest.loc[:,'Wilderness_Area1':'Soil_Type40']

Wilderness_data=forest.loc[:,'Wilderness_Area1': 'Wilderness_Area4']

Soil=forest.loc[:,'Soil_Type1':'Soil_Type40']
```

```python
cont_data
```

```python
def reverse_one_hot_Soil(row):
    for c in Soil.columns:
        if row[c]==1:
            return c  

cont_data['Soil_Type']=forest.apply(reverse_one_hot_Soil, axis=1)
```

```python
def reverse_one_hot(row):
    for c in Wilderness_data.columns:
        if row[c]==1:
            return c  

cont_data['Wilderness_Area']=forest.apply(reverse_one_hot, axis=1)
```

```python
cont_data['Cover_Type']=forest['Cover_Type']
```
